# Resize

```python
def plt_imshow(title, image):
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
image = cv2.imread(args["image"])
plt_imshow("Original", image)

r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))

# actual resizing of the image
resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt_imshow("Resized (Width)", resized)
```

```python
r = 50.0 / image.shape[0]
dim = (int(image.shape[1] * r), 50)

# resizing
resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt_imshow("Resized (Height)", resized)
```

```python
resized = imutils.resize(image, width=100)
plt_imshow("Resized via imutils", resized)
```

```python
methods = [
	("cv2.INTER_NEAREST", cv2.INTER_NEAREST),
	("cv2.INTER_LINEAR", cv2.INTER_LINEAR),
	("cv2.INTER_AREA", cv2.INTER_AREA),
	("cv2.INTER_CUBIC", cv2.INTER_CUBIC),
	("cv2.INTER_LANCZOS4", cv2.INTER_LANCZOS4)]


for (name, method) in methods: # Mengunakan metode interpolasi
	print("[INFO] {}".format(name))
	resized = imutils.resize(image, width=image.shape[1] * 3,
		inter=method)
	plt_imshow("Method: {}".format(name), resized)
```

# Rotate

```python
def plt_imshow(title, image):
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
image = cv2.imread(args["image"])
plt_imshow("Original", image)
(h, w) = image.shape[:2]
(cX, cY) = (w // 2, h // 2)

# rotate the image by 45 degrees around the center 
M = cv2.getRotationMatrix2D((cX, cY), 50, 0.8)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by 45 Degrees", rotated)

# rotate the image by -90 degrees 
M = cv2.getRotationMatrix2D((cX, cY), -90, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by -90 Degrees", rotated)
```

```python
# rotate the image around an arbitrary point rather than the center
M = cv2.getRotationMatrix2D((10, 15), 45, 1.1)
rotated = cv2.warpAffine(image, M, (w, h))
plt_imshow("Rotated by Arbitrary Point", rotated)
```

```python
# use imutils function to rotate the image by 180 degrees
rotated = imutils.rotate(image, 180)
plt_imshow("Dirotasikan 180 Derajat", rotated)
```

```python
# rotate the image by 25 degrees counterclockwise
rotated = imutils.rotate_bound(image, -25)
plt_imshow("Merotasikan tanpa di Crop", rotated)
```

# Translate

```python
def plt_imshow(title, image):
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python
image = cv2.imread(args["image"])
plt_imshow("Original", image)

M = np.float32([[1, 0, 10], [0, 1, 10]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt_imshow("Pojok kanan bawah", shifted)

M = np.float32([[1, 0, -10], [0, 1, 10]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt_imshow("Pojok kiri bawah", shifted)

shifted = imutils.translate(image, 0, 15)
plt_imshow("Kebawah", shifted)
```
